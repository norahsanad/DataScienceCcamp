````python
#import the needed library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
#define the dataframe
iris = pd.read_csv('https://raw.githubusercontent.com/DSomair/Masar_DS/main/Datasets/DVQuiz/Iris.csv')

#check if you get the data
print (iris)
````

**1 - Scatter plot x="SepalLengthCm", y="SepalWidthCm":**

````python
# Creating a scatter plot of Sepal Length vs Sepal Width
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'])  # Scatter plot
plt.xlabel('Sepal Length (cm)')  # Labeling x-axis
plt.ylabel('Sepal Width (cm)')   # Labeling y-axis
plt.title('Scatter Plot of Sepal Length vs Sepal Width')  # Adding title
plt.show()  # Displaying the plot
````

**2 - Scatter plot x="SepalLengthCm", y="SepalWidthCm" (Different color by species)**

````python
# Creating a scatter plot of Sepal Length vs Sepal Width with different colors for each species
sns.scatterplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species')  # Scatter plot with Seaborn
plt.title('Scatter Plot of Sepal Length vs Sepal Width (by Species)')  # Adding title
plt.show()  # Displaying the plot
````

**3- Boxplot for Petal Length for each species:**

````python
# Creating a boxplot of Petal Length by Species
sns.boxplot(data=iris, x='Species', y='PetalLengthCm')  # Boxplot with Seaborn
plt.title('Boxplot of Petal Length by Species')  # Adding title
plt.show()  # Displaying the plot
````

**4- Using Seaborn, try to plot kdeplot for the Petal length:**

````python
# Creating a violin plot of Petal Length by Species
sns.violinplot(x="Species", y="PetalLengthCm", data=iris)  # Violin plot with Seaborn
plt.xlabel('Species')  # Labeling x-axis
plt.ylabel('Petal Length (cm)')  # Labeling y-axis
plt.title('Violin Plot of Petal Length by Species')  # Adding title
plt.show()  # Displaying the plot
````

**5- Pairplot for all the variables (drop "Id") and make "hue" for each species:**

````python
# Creating a FacetGrid for KDE plot of Petal Length by Species
plt.figure(figsize=(10, 8))
# Creating a FacetGrid with hue as Species
# Mapping KDE plot onto the FacetGrid
# Adding legend to the plot
sns.FacetGrid(iris, hue="Species") \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
plt.xlabel('Petal Length (cm)') # Labeling x-axis
plt.title('KDE Plot of Petal Length by Species') # Adding title
plt.show() # Displaying the plot
````

**Another useful seaborn plot ( pairplot )**

````python
# Creating a pairplot of variables with hue for Species
sns.pairplot(iris.drop('Id', axis=1), hue='Species')  # Pairplot with Seaborn, dropping 'Id' column
plt.title('Pairplot of Variables (with hue for Species)')  # Adding title
plt.show()  # Displaying the plot
````

**boxplot() function in Pandas to create boxplots for each feature split out by species**

````python
# Grouping the data by 'Species'
grouped = iris.groupby('Species')

# Creating boxplots for each feature split out by species
boxplot = grouped.boxplot(figsize=(12, 8), layout=(1, 4))
plt.suptitle('Boxplot of Features by Species', fontsize=16)  # Adding a title for the overall plot
plt.tight_layout()  # Adjusting layout
plt.show()
````
